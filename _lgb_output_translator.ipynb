### LGBがp-scoreでoutputしてるので、f1にしてみる --
* 精度は微妙

```python
import pandas as pd
import numpy as np
from config import *
```

```python
train = pd.read_csv(data_path+"train.csv")
oof = pd.read_csv("output/tmp/oof.csv").rename(columns={"label":"prediction"})
oof = pd.merge(oof, train.loc[:, ["id", "label"]], how="left", on="id").rename(columns={"label":"label_true"})
```

```python
oof.head(2)
```

```python
def submission_translator(sub, threshold=0.5):
    sub["label"] = 0
    sub.loc[sub["prediction"]>threshold, "label"] = 1
    sub["label"] = sub["label"].astype(int)
    return sub
```

```python
from sklearn.metrics import accuracy_score, f1_score

acc_list = []
f1_list = []
for threshold in np.linspace(0, 1, 10):
    """ "prediction" -> "label" """
    oof = submission_translator(oof, threshold=threshold)

    acc_list.append(accuracy_score(oof["label"], oof["label_true"]))
    f1_list.append(f1_score(oof["label"], oof["label_true"]))
```

```python
log_df = pd.DataFrame({
    "threshold": list(np.linspace(0, 1, 10)),
    "acc": acc_list,
    "f1": f1_list
})
log_df
```
