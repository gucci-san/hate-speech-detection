### loss, f1, accの関係について実験

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
```

### acc vs f1, label=0から推移

```python
N = 10

clr_rgb = [(x/N, 0, 0) for x in range(N)]

df = pd.DataFrame({
    "y_true": [0]*N,
    "y_pred": [0]*N
})

f1_list = []
acc_list = []

for i in range(df.shape[0]):
    df.loc[i, "y_pred"] = 1
    f1_list.append(f1_score(df["y_true"], df["y_pred"]))
    acc_list.append(accuracy_score(df["y_true"], df["y_pred"]))

fig, ax = plt.subplots(1, 1, facecolor="white", figsize=(6, 4))
ax.scatter(f1_list, acc_list, c=clr_rgb)
ax.set_title("acc vs f1, with label 0 to 1", fontsize=18)
ax.set_xlabel("f1_score[-]", fontsize=14)
ax.set_ylabel("acc_score[-]", fontsize=14)

ax.set_xlim([-0.1, 1.1]); ax.set_ylim([-0.1, 1.1])
```

### acc vs f1, label=1から推移

```python
N = 10

clr_rgb = [(x/N, 0, 0) for x in range(N)]

df = pd.DataFrame({
    "y_true": [1]*N,
    "y_pred": [1]*N
})

f1_list = []
acc_list = []

for i in range(df.shape[0]):
    df.loc[i, "y_pred"] = 0
    f1_list.append(f1_score(df["y_true"], df["y_pred"]))
    acc_list.append(accuracy_score(df["y_true"], df["y_pred"]))

fig, ax = plt.subplots(1, 1, facecolor="white", figsize=(6, 4))
ax.scatter(f1_list, acc_list, c=clr_rgb)
ax.set_title("acc vs f1, with label 1 to 0", fontsize=18)
ax.set_xlabel("f1_score[-]", fontsize=14)
ax.set_ylabel("acc_score[-]", fontsize=14)

ax.set_xlim([-0.1, 1.1]); ax.set_ylim([-0.1, 1.1])
```

### pre vs rec, label=0から推移

```python
N = 10

clr_rgb = [(x/N, 0, 0) for x in range(N)]

df = pd.DataFrame({
    "y_true": [0]*N,
    "y_pred": [0]*N
})

pre_list = []
rec_list = []

for i in range(df.shape[0]):
    df.loc[i, "y_pred"] = 1
    pre_list.append(precision_score(df["y_true"], df["y_pred"]))
    rec_list.append(recall_score(df["y_true"], df["y_pred"]))

fig, ax = plt.subplots(1, 1, facecolor="white", figsize=(6, 4))
ax.scatter(pre_list, rec_list, c=clr_rgb)
ax.set_title("pre vs rec, with label 0 to 1", fontsize=18)
ax.set_xlabel("precision [-]", fontsize=14)
ax.set_ylabel("recall [-]", fontsize=14)

ax.set_xlim([-0.1, 1.1]); ax.set_ylim([-0.1, 1.1])
```

```python
N = 10

clr_rgb = [(x/N, 0, 0) for x in range(N)]

df = pd.DataFrame({
    "y_true": [1]*N,
    "y_pred": [1]*N
})

pre_list = []
rec_list = []

for i in range(df.shape[0]):
    df.loc[i, "y_pred"] = 0
    pre_list.append(precision_score(df["y_true"], df["y_pred"]))
    rec_list.append(recall_score(df["y_true"], df["y_pred"]))

fig, ax = plt.subplots(1, 1, facecolor="white", figsize=(6, 4))
ax.scatter(pre_list, rec_list, c=clr_rgb)
ax.set_title("pre vs rec, with label 1 to 0", fontsize=18)
ax.set_xlabel("precision [-]", fontsize=14)
ax.set_ylabel("recall [-]", fontsize=14)

ax.set_xlim([-0.1, 1.1]); ax.set_ylim([-0.1, 1.1])
```

つまり
* 全部0 ... (pre, rec) = (0, 0)
* 全部1 ... (pre, rec) = (1, 0.9)
